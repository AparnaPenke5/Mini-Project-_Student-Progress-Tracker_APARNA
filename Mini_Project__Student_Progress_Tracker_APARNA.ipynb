MINI PROJECT

```


while True:

    print("-----------------------------------------")
    print("STUDENT PROGRESS TRACKER SYSTEM")
    print("-----------------------------------------")


    name = input("Enter Student Name: ")
    student_class = input("Enter Class: ")
    roll = input("Enter Roll Number: ")


    subjects = int(input("\nHow many subjects do you want to enter? "))

    total = 0
    fail_flag = False

    for i in range(1, subjects + 1):
        mark = float(input(f"Enter marks for subject {i}: "))
        total += mark
        if mark < 35:
            fail_flag = True

    average = total / subjects


    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 50:
        grade = "C"
    elif average >= 35:
        grade = "D"
    else:
        grade = "Fail"


    if fail_flag or grade == "Fail":
        result = "FAIL"
    else:
        result = "PASS"


    print("\n-----------------------------------------")
    print("        STUDENT PROGRESS REPORT")
    print("-----------------------------------------")
    print(f"Name         : {name}")
    print(f"Class        : {student_class}")
    print(f"Roll Number  : {roll}")
    print(f"Total Marks  : {total}")
    print(f"Average      : {average:.2f}")
    print(f"Grade        : {grade}")
    print(f"Result       : {result}")
    print("-----------------------------------------")


    while True:
        print("\nWhat do you want to do next?")
        print("1. Enter marks again")
        print("2. View report again")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            break
        elif choice == "2":

            print("\n-----------------------------------------")
            print("        STUDENT PROGRESS REPORT")
            print("-----------------------------------------")
            print(f"Name         : {name}")
            print(f"Class        : {student_class}")
            print(f"Roll Number  : {roll}")
            print(f"Total Marks  : {total}")
            print(f"Average      : {average:.2f}")
            print(f"Grade        : {grade}")
            print(f"Result       : {result}")
            print("-----------------------------------------")
        elif choice == "3":
            print("\nExiting Program... Goodbye!")
            exit()
        else:
            print("Invalid choice! Try again.")
```
